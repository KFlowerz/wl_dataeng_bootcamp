```python
import pandas as pd
        
print("Environment Ready")
```

```python
df = pd.read_csv('Data/profeco/sample.csv')
df.head()
```

```python
df.describe()
```

```python
num_unique_comch= df['cadenaComercial'].nunique();
print('How many comercial chanins are monitored and, therefore, included in this database?')
print(num_unique_comch)
```

```python
df_pp = df.loc[:,['estado','producto','precio']]
df_top = df_pp.groupby(['estado','producto']).sum()['precio']
print('Summary of prices per product')
df_top
```

```python
top_productos = df.groupby('cadenaComercial')['producto'].nunique()
print('Summary of number of products per commercial chain:')
top_productos
```

```python
top_ten_prod = df.assign(sortkey = df.groupby('producto')['precio'].transform('sum'))\
  .sort_values(['sortkey','producto','precio'], ascending=[False,True,False])\
  .set_index(['producto'])\
  .drop('sortkey', axis=1)

print(top_ten_prod.head(10))
```
